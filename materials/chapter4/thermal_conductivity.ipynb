```python
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

```python
_cu_data = "https://colab.research.google.com/github/hnakata-JP/seminar_Ekin/blob/main/materials/chapter4/cu_thermal_conductivity_interpolated.csv"
df = pd.read_csv(_cu_data)
# print(df.head())
_other_data = "https://colab.research.google.com/github/hnakata-JP/seminar_Ekin/blob/main/materials/chapter4/other_thermal_conductivity.csv"
df1 = pd.read_csv(_other_data)
# print(df1.head())
```

```python
Cu = ['RRR10', 'RRR20', 'RRR30', 'RRR40', 'RRR60', 'RRR80', 'RRR100', 'RRR300']
color_list = ['gold', 'orange', 'sandybrown', 'coral', 'tomato', 'orangered', 'red']
T = df['T[K]']
ax = plt.axes()
plt.rcParams['font.size'] = 16
for _i in range(1, len(Cu)):
    K = df[f'{Cu[_i]}']
    ax.plot(T, K, label=f"Cu_{Cu[_i]}", color=color_list[_i-1])

ax.set_xlabel('Temperature (K)')
ax.set_ylabel('Thermal Conductivity (W/(m$\cdot$K))')

handles, labels =  ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], bbox_to_anchor=(1.05, 1), loc='upper left')

ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(which='both', linestyle='--', linewidth=0.5)
```

```python
materials = ['N2_Gas', 'He_Gas', 'Pyrex', 'G10', 'Manganin', 'Constantan', 'Al6061-T6', 'Al', 'SS304L_RRR30']
ax = plt.axes()
plt.rcParams['font.size'] = 16
for _m in materials:
    _T = df1[f'T_{_m}']
    _K = df1[f'K_{_m}']
    ax.plot(_T, _K, label=f'{_m}')

ax.set_xlabel('Temperature (K)')
ax.set_ylabel('Thermal Conductivity (W/(m$\cdot$K))')

handles, labels =  ax.get_legend_handles_labels()
ax.legend(handles[::-1], labels[::-1], bbox_to_anchor=(1.05, 1), loc='upper left')

ax.set_xscale('log')
ax.set_yscale('log')
ax.grid(which='both', linestyle='--', linewidth=0.5)
```
